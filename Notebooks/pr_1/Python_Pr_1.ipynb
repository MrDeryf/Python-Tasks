3.4. (уровень сложности: низкий)

Некто попытался реализовать «наивную» функцию умножения с помощью сложений. К сожалению, в коде много ошибок. Сможете ли вы их исправить?

Добавьте к naive_mul автоматическое тестирование на случайных данных. Сравнивайте с встроенным умножением, используя конструкцию assert.


```python
import random


def naive_mul(x, y):
    out = 0
    if x * y > 0:
        sign = 1
    else:
        sign = -1
    x = abs(x)
    y = abs(y)
    for i in range(y):
        out += x
    return sign * out


x, y = random.randint(-100, 100), random.randint(-100, 100)
assert naive_mul(x, y) == x * y, f"У тебя {naive_mul(x, y)}, надо {x * y}"
```

3.5. (уровень сложности: средний)

Реализуйте функцию fast_mul в соответствии с алгоритмом двоичного умножения в столбик (без рекурсии!). Добавьте автоматическое тестирование, как в случае с naive_mul.

```python
def fast_mul(x, y):
    out = 0
    if x * y > 0:
        sign = 1
    else:
        sign = -1
    x = abs(x)
    y = abs(y)
    while x >= 1:
        if x % 2 != 0:
            out += y
        x //= 2
        y *= 2
    return sign * out


x, y = random.randint(-100, 100), random.randint(-100, 100)
assert fast_mul(x, y) == x * y, f"У тебя {fast_mul(x, y)}, надо {x * y}"
```

3.6. (уровень сложности: средний)

Реализуйте аналогичную функцию fast_pow для возведения в степень. Решение необходимо получить только с помощью небольших модификаций предыдущего решения.

```python
def fast_pow(x, y):
    out = 1
    sign = 1
    if y < 0:
        sign = -1
    y = abs(y)
    for _ in range(y):
        out = fast_mul(out, x)
    return out**sign


x, y = random.randint(-100, 100), random.randint(-100, 100)
assert fast_pow(x, y) == x**y, f"Дано: {x, y}. У тебя {fast_pow(x, y)}, надо {x**y}"
```

3.7. (уровень сложности: средний)

Предположим, что стоит задача перемножить два 16-битных числа без знака. Однако в наличии у нас имеется только операция умножения 8-битных чисел. Оказывается, можно применить четыре 8-битных умножения, если разбить исходные числа на 8-битные половинки. Побитовые сдвиги умножениями не считаются.

Выведите формулу для умножения «половинок» и реализуйте функцию mul16(x, y), используя приведенную ниже функцию mul_bits, которая имитирует умножение с ограниченной числом bits разрядностью аргументов.


```python
def mul_8(x, y):
    x &= 2**8 - 1
    y &= 2**8 - 1
    return x * y


def mul_16(x, y):
    return (
        mul_8(x, y)
        + (mul_8(x // 256, y) * 256 + mul_8(x, y // 256) * 256)
        + mul_8(x // 256, y // 256) * 2**16
    )
```

3.8. (уровень сложности: средний)

Четыре умножения из предыдущей задачи – не предел. Советский ученый Анатолий Алексеевич Карацуба в 1960 г. предложил формулу (а для общего случая – рекурсивный алгоритм), требующую лишь трех умножений. Выведите эту формулу с подсказками от преподавателя и реализуйте функцию mul16k(x, y).


```python
 def mul_16(x, y):
     a_c = mul_8(x // 256, y // 256)
     b_d = mul_8(x, y)
     smart = mul_8(x // 256 + x % 256, y // 256 + y % 256)
     return a_c * 2 ** 16 + (smart - a_c - b_d) * 256 + b_d

print(mul_16(5, 10))
```

Ваша функция должна выдать текст функции f(x) (умножение на ранее заданный y), тело которой состоит из некоторого числа присваиваний. Для вывода функции используйте print.
```python
def fast_mul_gen(y):
    assignment = "    out = out + x\n"
    out = f"def f_{y}(x):\n    out = x\n"
    for _ in range(y - 1):
        out += assignment
    out += "    return out"
    return out


print(fast_mul_gen(random.randint(1, 10)))
```

```python
import math
import tkinter as tk


def draw(shader, width, height):
    image = bytearray((0, 0, 0) * width * height)
    for y in range(height):
        for x in range(width):
            pos = (width * y + x) * 3
            color = shader(x / width, y / height)
            normalized = [max(min(int(c * 255), 255), 0) for c in color]
            image[pos : pos + 3] = normalized
    header = bytes(f"P6\n{width} {height}\n255\n", "ascii")
    return header + image


def main(shader):
    label = tk.Label()
    img = tk.PhotoImage(data=draw(shader, 256, 256)).zoom(2, 2)
    label.pack()
    label.config(image=img)
    tk.mainloop()


def shader(x, y):
    center_r_x, center_r_y = 0.5, 0.5
    center_yg_x, center_yg_y = 0.4, 0.4
    radius_yg = 0.5
    distance_yg = math.sqrt((x - center_yg_x) ** 2 + (y - center_yg_y) ** 2)
    radius_r = 0.5
    distance_r = math.sqrt((x - center_r_x) ** 2 + (y - center_r_y) ** 2)
    if distance_yg < radius_yg:
        if distance_r > radius_r:
            q = radius_yg - distance_yg
            return 0, 0, 0
        else:
            normalized_dist1 = distance_yg / radius_yg
            normalized_dist2 = distance_r / radius_r
            # Вычисляем коэффициенты для смешивания цветов
            green_coeff = 1 - normalized_dist1
            red_coeff = 1 - normalized_dist2
            # Вычисляем смешанный цвет
            yellow = (1, 1, 0)  # Желтый
            green = (0, 1, 0)  # Зеленый
            red = (1, 0, 0)  # Красный
            black = (0, 0, 0)
            mixed_color = (
                green_coeff * green[0] + red_coeff * red[0],
                green_coeff * green[1] + red_coeff * red[1],
                green_coeff * green[2] + red_coeff * red[2],
            )
            t = normalized_dist1 + normalized_dist2
            return tuple(c1 * (1 - t) + c2 * t for c1, c2 in zip(yellow, mixed_color))

    else:
        return 0, 0, 0


main(shader)
```

4.3. (уровень сложности: средний)

Изобразите знаменитого персонажа компьютерной игры.

```python
import math
import tkinter as tk


def draw(shader, width, height):
    image = bytearray((0, 0, 0) * width * height)
    for y in range(height):
        for x in range(width):
            pos = (width * y + x) * 3
            color = shader(x / width, y / height)
            normalized = [max(min(int(c * 255), 255), 0) for c in color]
            image[pos : pos + 3] = normalized
    header = bytes(f"P6\n{width} {height}\n255\n", "ascii")
    return header + image


def main(shader):
    label = tk.Label()
    img = tk.PhotoImage(data=draw(shader, 256, 256)).zoom(2, 2)
    label.pack()
    label.config(image=img)
    tk.mainloop()


def shader(x, y):
    center_x = 0.5
    center_y = 0.5
    radius = 0.4

    center_mouth_x = 0.65
    center_mouth_y = 0.25
    radius_mouth = 0.06

    distance_to_center = math.sqrt((x - center_x) ** 2 + (y - center_y) ** 2)
    distance_to_center_mouth = math.sqrt(
        (x - center_mouth_x) ** 2 + (y - center_mouth_y) ** 2
    )

    angle = math.atan2(y - center_y, x - center_x)
    angle_start = -math.pi / 6
    angle_end = math.pi / 6

    if distance_to_center <= radius:
        if (
            angle_start <= angle <= angle_end
            or distance_to_center_mouth <= radius_mouth
        ):
            return 0, 0, 0
        else:
            return 1, 1, 0
    else:
        return 0, 0, 0


main(shader)
```
