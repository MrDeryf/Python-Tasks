```python
def encode_val(length, rep, bin_num):
    return bin(
        int("_".join(str(bin(bin_num)).lstrip("0b")[i] * rep for i in range(length)), 2)
    )


def decode_val(length, rep, num):
    bin_num = str(bin(num)).lstrip("0b")
    if len(bin_num) < length:
        bin_num = "0" * (length - len(bin_num)) + bin_num
    out = [
        "1" if bin_num[i : i + rep].count("1") > rep // 2 else "0"
        for i in range(0, length, rep)
    ]
    return bin(int("".join(out), 2))


encrypted_message = [
    815608,
    2064837,
    2093080,
    2063879,
    196608,
    2067983,
    10457031,
    1830912,
    2067455,
    2093116,
    1044928,
    2064407,
    6262776,
    2027968,
    4423680,
    2068231,
    2068474,
    1999352,
    1019903,
    2093113,
    2068439,
    2064455,
    1831360,
    1936903,
    2067967,
    2068456,
]
print([bin(x) for x in encrypted_message])  # 3 * 8
for i in range(len(encrypted_message)):
    encrypted_message[i] = decode_val(24, 3, encrypted_message[i])
print(encrypted_message)
for x in encrypted_message:
    print(chr(int(x, 2)), end="")
```

5.3. (уровень сложности: низкий)

Советский и российский математик Владимир Иосифович Левенштейн в 1965 г. предложил свой вариант редакционного расстояния. В расстоянии Левенштейна учитываются не только замены символов, но также операции вставки и удаления. То есть, в отличие от расстояния Хэмминга, с помощью расстояния Левенштейна можно сравнивать последовательности с разными длинами.



Реализуйте функцию вычисления расстояния Левенштейна.

```python
from functools import cache


@cache
def lev_dist(a, b, i=-1, j=-1):
    if i == -len(a) - 1 or j == -len(b) - 1:
        return max(len(a) + i + 1, len(b) + j + 1)
    return min(
        lev_dist(a, b, i - 1, j) + 1,
        lev_dist(a, b, i, j - 1) + 1,
        lev_dist(a, b, i - 1, j - 1) + (a[i] != b[j]),
    )


print(lev_dist("напесал", "по-моему"))
```

5.4. (уровень сложности: средний)

Реализованная функция вычисления расстояния Левенштейна дает только миниальное число операций. Было бы полезно узнать, какие именно операции произведены над исходной строкой.

```python
from functools import cache


@cache
def lev_dist_ops(a, b, i=-1, j=-1):
    if i == -len(a) - 1 or j == -len(b) - 1:
        if len(a) + i + 1 > len(b) + j + 1:
            return ["delete" for _ in range(len(a) + i + 1)]
        else:
            return ["insert" for _ in range(len(b) + j + 1)]
    delete = lev_dist_ops(a, b, i - 1, j) + ["delete"]
    insert = lev_dist_ops(a, b, i, j - 1) + ["insert"]
    sub = lev_dist_ops(a, b, i - 1, j - 1) + (["sub"] if a[i] != b[j] else [])
    return min(delete, insert, sub, key=len)


print(lev_dist_ops("столб", "слон"))
```

Это может помочь компактно хранить данные – для измененной версии файла мы храним только команды, которые производят эти изменения. В этой задаче необходимо сгенерировать код на Питоне, который из исходного списка создает список-результат с помощью минимального числа операций вставки, замены и удаления.

```python
from functools import cache
from random import randint


@cache
def lev_dist_ops(a, b, i=-1, j=-1):
    if i == -len(a) - 1 or j == -len(b) - 1:
        if len(a) + i + 1 > len(b) + j + 1:
            return [("del", k, len(b) + j) for k in range(len(a) + i + 1)]
        else:
            return [("insert", len(a) + i, k) for k in range(len(b) + j + 1)]
    delete = lev_dist_ops(a, b, i - 1, j) + [("del", len(a) + i, len(b) + j)]
    insert = lev_dist_ops(a, b, i, j - 1) + [("insert", len(a) + i, len(b) + j)]
    sub = lev_dist_ops(a, b, i - 1, j - 1) + (
        [("sub", len(a) + i, len(b) + j)] if a[i] != b[j] else []
    )
    return min(delete, insert, sub, key=len)


def lev_dist_gen(a, b):
    ops = lev_dist_ops(a, b)
    code = []
    offset = 0
    for op in ops:
        match op[0]:
            case "del":
                code.append(f"del x[{op[1] + offset}]")
                offset -= 1
            case "insert":
                code.append(f"x.insert({op[1] + offset + 1}, y[{op[2]}])")
                offset += 1
            case "sub":
                code.append(f"x[{op[1] + offset}] = y[{op[2]}]")
    return code


for _ in range(1000):
    x = list(chr(randint(97, 122)) for _ in range(randint(3, 5)))
    y = list(chr(randint(97, 122)) for _ in range(randint(3, 5)))
    #    print(x, y)
    script = "\n".join(lev_dist_gen("".join(x), "".join(y)))
    #    print(script)
    exec(script)
    assert x == y, f"{x}, {y}"
```

В то же время, контекст цифровой трансформации повышает вероятность несанкционированной кастомизации волатильных активов. Тем не менее, экспоненциальный рост Big Data обостряет проблему синергетического эффекта государственно-частных партнёрств.


```python
from random import shuffle

text = """Коллеги, прагматичный подход к цифровым платформам открывает новые возможности для универсальной коммодитизации нежелательных последствий. Следовательно, ускорение блокчейн-транзакций выдвигает новые требования бюджетного финансирования непроверенных гипотез. Соответственно, программа прорывных исследований несёт в себе риски компрометации конфиденциальных цифровых следов граждан.

С другой стороны, парадигма цифровой экономики не оставляет шанса для нормативного регулирования знаний и компетенций. Вместе с тем, диджитализация бизнес-процессов заставляет искать варианты дальнейшего углубления опасных экспериментов. Однако, совокупность сквозных технологий расширяет горизонты практического применения внезапных открытий.

В то же время, контекст цифровой трансформации повышает вероятность несанкционированной кастомизации волатильных активов. Тем не менее, экспоненциальный рост Big Data обостряет проблему синергетического эффекта государственно-частных партнёрств."""

table = [[], [], [], [], []]
for sentence in text.split("."):
    words = sentence.split()
    split_points = [-1, -1, -1, -1]
    for i in range(len(words)):
        if words[i].endswith(","):
            split_points[0] = i
        elif (
            split_points[1] == -1
            and split_points[0] >= 0
            and (words[i].endswith("ет") or words[i].endswith("ёт") or words[i] == "не")
        ):
            split_points[1] = i
        elif (
            split_points[2] == -1
            and split_points[1] >= 0
            and (
                words[i].endswith("ого")
                or words[i].endswith("его")
                or words[i].endswith("ии")
                or words[i].endswith("ой")
            )
        ):
            split_points[2] = i
        elif (
            split_points[3] == -1
            and split_points[2] >= 0
            and (words[i].endswith("ий") or words[i].endswith("ых"))
        ):
            split_points[3] = i
    if split_points != [-1, -1, -1, -1]:
        table[0].append(" ".join(words[: split_points[0] + 1]))
        table[1].append(" ".join(words[split_points[0] + 1 : split_points[1]]))
        table[2].append(" ".join(words[split_points[1] : split_points[2]]))
        table[3].append(" ".join(words[split_points[2] : split_points[3]]))
        table[4].append(" ".join(words[split_points[3] :]) + ".")
print(table)

pattern = [list(range(len(table[0]))) for _ in range(len(table))]
for seq in pattern:
    shuffle(seq)
for i in range(len(table[0])):
    if pattern[0][i] == 0:
        pattern[0][0], pattern[0][i] = pattern[0][i], pattern[0][0]
sentences = []
for i in range(len(pattern[0])):
    sentences.append(
        " ".join([str(table[j][pattern[j][i]]) for j in range(len(pattern))])
    )
for i in range(0, len(sentences), 3):
    print(" ".join(sentences[i : i + 3]))
    print("\n")
```
